```python
import os
from firecrawl import Firecrawl
from IPython.display import display,Markdown,JSON,HTML
from dotenv import load_dotenv
import pandas as pd

load_dotenv()

app = Firecrawl(api_key=os.getenv("FIRECRAWL_API_KEY"))

```

# scrape

```python
# Scrape a website in markdown format:
doc = app.scrape("https://medium.com/p/7c198a15772b", formats=["markdown", "html"])
display(Markdown(doc.markdown))
```

```python
# Scrape a website in markdown format:
doc = app.scrape("https://medium.com/@thanseefpp/effortless-ocr-implementation-comprehensive-guide-with-pytesseract-keras-ocr-and-easyocr-code-f45163b7ea93", formats=["markdown", "html"])
display(Markdown(doc.markdown))
```

```python
# Scrape a website in json format:
from pydantic import BaseModel


class CompanyInfo(BaseModel):
    company_mission: str
    supports_sso: bool
    is_open_source: bool
    is_in_yc: bool

result = app.scrape(
    'https://medium.com/@thanseefpp/effortless-ocr-implementation-comprehensive-guide-with-pytesseract-keras-ocr-and-easyocr-code-f45163b7ea93',
    formats=[{
      "type": "json",
      "schema": CompanyInfo.model_json_schema()
    }],
    only_main_content=False,
    timeout=120000
)

display(JSON((result.json)))
```

```python
# Scrape a website in json format:
result = app.scrape(
    'https://medium.com/p/7c198a15772b',
    formats=[{
      "type": "json",
      "prompt": "Extract the company mission from the page."
    }],
    only_main_content=False,
    timeout=120000
)

# print(result.json)
display(JSON(result.json))
```

```python
doc = app.scrape(
    "https://medium.com/p/7c198a15772b",
    formats=[
        "html",
        {
            "type": "json",
            "prompt": "Extract the article title, author, and main content.",
            "schema": {
                "title": "string",
                "author": "string",
                "content": "string"
            }
        }
    ]
)
display(doc)
```

```python

# Scrape a website in html format:
doc = app.scrape("https://www.uefa.com/uefachampionsleague/", formats=["markdown", "html"])
display(HTML(doc.html))
df= pd.DataFrame(doc)
df
```
