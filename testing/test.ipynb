```python
from firecrawl import Firecrawl
from pydantic import BaseModel

app = Firecrawl(api_key="fc-1d8628ff311b4202b90ce0db78100bea")

class CompanyInfo(BaseModel):
    company_mission: str
    supports_sso: bool
    is_open_source: bool
    is_in_yc: bool

result = app.scrape(
    'https://medium.com/netflix-techblog/supercharging-the-ml-and-ai-development-experience-at-netflix-b2d5b95c63eb',
    formats=[{
      "type": "json",
      "schema": CompanyInfo.model_json_schema()
    }],
    only_main_content=False,
    timeout=120000
)

print(result)
```

```python
def scrape_website(url: str):
    try:
        result = Firecrawl.scrape(url, formats=['json'])
        
        if hasattr(result, 'markdown') and hasattr(result, 'html'):
            return {
                'json':result.json
                # 'markdown':result.markdown,
                # 'html':result.html
            }
        else:
            return result
    except:
        print(f'Failed to scrape url {url}')
```
